5. What was the largest change between any two days (based on Closing Price)?

```python
close = [row['Close'] for row in list_of_dict]
```

```python
def difference(dataset, interval=1):
    diff = list()
    for i in range(interval, len(dataset)):
        value = dataset[i] - dataset[i - interval]
        diff.append(value)
    return diff
```

```python
diff_close = difference(close,1)
```

```python
print('The largest change between any two days based on closing price was ', max(diff_close))
```

```python
def manual_median(seq_list):
    # seq_list takes in a list of values
    n_num = seq_list
    n = len(n_num)
    n_num = sorted(n_num)
  
    if n % 2 == 0:
        median1 = n_num[n//2]
        median2 = n_num[n//2 - 1]
        median = (median1 + median2)/2
    else:
        median = n_num[n//2]
    print("Median is: " + str(median))
```
